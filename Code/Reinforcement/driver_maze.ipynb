## **Import Libraries**

```
from IPython.display import HTML
import maze
import matplotlib.pyplot as plt
import numpy as np
```

## **1. Create Model**

```
width = 5
height = 5
epsilon = 0.0
model = maze.maze(width,height,epsilon)
```

## **2. Simulate**

```
nepisode = 80
alpha = 1
model.qlearning(nepisode,alpha)
```

## **3. Plot results**
